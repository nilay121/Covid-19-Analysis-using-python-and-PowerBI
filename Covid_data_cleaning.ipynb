```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

**Working with the Data only for India**

```python
#New DataFrame containing only values related to India
df_new=df[df['iso_code']=="IND"].reset_index().drop(columns=['index'])
```

```python
df_new.shape
```

```python
print(f"The starting date for the data is is :-{df_new['date'].min()}")
print(f"The end date for the data is is :-{df_new['date'].max()}")
```

```python
plt.figure(figsize=(20,7))
df_new.isnull().sum().plot(kind='bar')
```

```python
df_new
```

```python
#Creating the final DataFrame having appropriate columns
data={
    'date':df_new['date'],
    'total_cases':df_new['total_cases'],
    'new_cases':df_new['new_cases'],
    'total_deaths':df_new['total_deaths'],
    'new_deaths':df_new['new_deaths'],
    'total_tests':df_new['total_tests'],
    'new_test':df_new['new_tests'],
    'positive_rate':df_new['positive_rate'],
    'total_vaccinations':df_new['total_vaccinations'],
    'new_vaccinations':df_new['new_vaccinations'],
    'people_vaccinated':df_new['people_vaccinated']
}
df_final=pd.DataFrame(data)
```
